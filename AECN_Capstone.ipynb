```python
#import dependencies

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in csv files

percapita_income = pd.read_csv("PerCapitaIncome_float.csv", index_col=0)
child_dependency_ratio = pd.read_csv("child_dependency_ratio.csv", index_col=0)
families = pd.read_csv("families_float.csv", index_col=0)
commute = pd.read_csv("commute_time.csv", index_col=0)
educational_attainment = pd.read_csv("educational_attainment_float.csv", index_col=0)
```

```python
#merge csv files into new dataframe w/concat and transpose the data

eastern_mountain_region = pd.concat([families, commute, percapita_income, child_dependency_ratio]).T
```

```python
#view new dataframe 

eastern_mountain_region
```

```python
#strip columns of empty space. 
eastern_mountain_region.columns = ["Total Families", "Mean Travel Time to Work in Minutes", "Per Capita Income", "Child Dependency Ratio"]
```

```python
#use matplotlib to compare total # of families to per capita income in the eastern mountain region counties
#Need to improve visual: How can I label the points on the scatter plot as each county name?

plt.scatter(eastern_mountain_region["Total Families"], eastern_mountain_region["Per Capita Income"])
plt.show()
```

```python
#scatterplot to compare per capita income against work commue for eastern mountain counties

plt.scatter(eastern_mountain_region["Per Capita Income"], eastern_mountain_region["Mean Travel Time to Work in Minutes"])
plt.show
```

```python
#make scatter plot w/ matplotlib comparing families and child dependency ratio

plt.scatter(eastern_mountain_region["Total Families"], eastern_mountain_region["Child Dependency Ratio"])
plt.show
```
